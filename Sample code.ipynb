```python
import pandas as pd
import numpy as np
import math
# Method 1: Linear transformation.

# non-transformated data. (data missing is allowed)
panel=pd.read_excel("simulation data.xlsx")
# A custom substring function for seeking normal range location
def isSubstring(s1,s2,s):                            
    tag = False
    if s.find(s1) != -1 or s.find(s2) != -1: 
        tag = True
    return tag
# Range_loc: Store normal range location.
Range_loc=[]                                         
name=panel.columns.values
str1=str('lower')
str2=str('upper')
for i in range(len(name)):
    if isSubstring(str1,str2,name[i])== True:
        Range_loc.append(i)
    
panel_lineartrans=panel.copy(deep=True).astype('float') 
# number_of_items: the number of test item in the panel.
number_of_items=int(len(Range_loc)/2) 

for i in range(number_of_items):
    table=panel.iloc[:,[2*i+number_of_items,2*i+number_of_items+1]].value_counts()
    Number_of_ranges=len(table)
    # index[[i]]: Record the position where the i-th normal range is used.
    index=[[]for z in range(Number_of_ranges)]
    # Complete Linear transformation.
    for j in range(Number_of_ranges):                                
        index[j]=np.where(np.sum(panel.iloc[:,[2*i+number_of_items,2*i+number_of_items+1]]==table.index[j],axis=1)==2)[0].tolist() #The position of each normal range of each test item
        panel_lineartrans.iloc[index[j],i]=(panel_lineartrans.iloc[index[j],i]-table.index[j][0])*(table.index[0][1]-table.index[0][0])/(table.index[j][1]-table.index[j][0])+table.index[0][0]        
    # Change the normal range used.
    panel_lineartrans.iloc[:,2*i+number_of_items]=table.index[0][0] 
    panel_lineartrans.iloc[:,2*i+number_of_items+1]=table.index[0][1]  
    
#linear-transformated result

panel_lineartrans
```

```python
# Method 2: ECDF transformation

# non-transformated data. (data missing is allowed)
panel=pd.read_excel("simulation data.xlsx")

# A custom substring function for seeking normal range location
def isSubstring(s1,s2,s):                            
    tag = False
    if s.find(s1) != -1 or s.find(s2) != -1: 
        tag = True
    return tag
# Range_loc: Store normal range location.
Range_loc=[]                                         
name=panel.columns.values
str1=str('lower')
str2=str('upper')
for i in range(len(name)):
    if isSubstring(str1,str2,name[i])== True:
        Range_loc.append(i)
    

# number_of_items: the number of test item in the panel.
number_of_items=int(len(Range_loc)/2) 

panel_ecdf=panel.copy(deep=True).astype('float')  

# threshold_set: Minimum allowed number of samples.
threshold_set=100
for i in range(number_of_items):
    table=panel.iloc[:,[2*i+number_of_items,2*i+number_of_items+1]].value_counts()
    Number_of_ranges=len(table)
    #Enough samples to participate in ECDF calculation.
    threshold=len(table[table.values>threshold_set])                  
    index=[[]for z in range(Number_of_ranges)]
    # ECDF transformation also needs Linear transformation.
    for j in range(Number_of_ranges):                                
        index[j]=np.where(np.sum(panel.iloc[:,[2*i+number_of_items,2*i+number_of_items+1]]==table.index[j],axis=1)==2)[0].tolist()
        panel_ecdf.iloc[index[j],i]=(panel_ecdf.iloc[index[j],i]-table.index[j][0])*(table.index[0][1]-table.index[0][0])/(table.index[j][1]-table.index[j][0])+table.index[0][0]        
    panel_ecdf.iloc[:,2*i+number_of_items]=table.index[0][0] 
    panel_ecdf.iloc[:,2*i+number_of_items+1]=table.index[0][1]  
    if threshold>1:
        n=[[]for z in range(threshold)]                   #The number of occurrences of each normal range
        n1=[[]for z in range(threshold)]                  #The number of occurrences below the lower bound in each normal range
        n2=[[]for z in range(threshold)]                  #The number of occurrences above the upper bound in each normal range
        F=[[]for z in range(threshold)]                   #Quantile obtained by maximum likelihood
        P1=[[]for z in range(threshold)]                  #The proportion of the test conclusion below the normal value under different normal range
        P2=[[]for z in range(threshold)]                  #The proportion of the test conclusion above the normal value under different normal range
        for j in range(threshold):                       
            n[j]=table.values[j]
            n1[j]=np.sum(panel.iloc[index[j],i]<=table.index[j][0])
            n2[j]=np.sum(panel.iloc[index[j],i]>table.index[j][1])
            P1[j]=np.sum(panel.iloc[index[j],i]<=table.index[j][0])/len(panel.iloc[index[j],i])
            P2[j]=np.sum(panel.iloc[index[j],i]>table.index[j][1])/len(panel.iloc[index[j],i])
#If there is a large difference between the proportions under different normal ranges, it will not participate in the ECDF calculation
        if(all(P1) or not any(P1))and (all(P2) or not any(P2)):  
            x=np.sum(n1)/np.sum(n)
            y=np.sum(n2)/np.sum(n)
            for j in range(threshold):
                rank=panel_ecdf.iloc[index[j],i].rank().tolist()      #Ranking of observations under j-th normal range
                F[j]=[[0]for u in range(n[j])]                        #Empirical cumulative density
                for k in range(n[j]):
                    if rank[k]<=n1[j]:
                        F[j][k]=rank[k]*x/n1[j]
                    if rank[k]>n1[j] and rank[k]<=n[j]-n2[j]:
                        F[j][k]=(rank[k]-n1[j])*(1-x-y)/(n[j]-n1[j]-n2[j])+x
                    if rank[k]>n[j]-n2[j]:
                        F[j][k]=(rank[k]-(n[j]-n2[j]))*(y/n2[j])+1-y       
                    if j!=0:
                        panel_ecdf.iloc[index[j][k],i]=panel_ecdf.iloc[index[0][np.argmin(abs(F[j][k]-F[0]))],i]  #Complete ECDF transformation

#ECDF-transformated result


panel_ecdf
```
